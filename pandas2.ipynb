```python
#fas-정재환
import pandas as pd
```

```python
#저장된 csv파일을 0번째줄을 인덱스 시작으로 읽어온다
data_health = pd.read_csv("health.csv", index_col=0)
data_health = data_health.drop(['NUM_CODE'], axis=1)
data_health
```

```python
#역순으로 하단에서 50개를 차례대로 출력
data_health.loc[::-1].head(10)
```

```python
#index순서를 반대로 뒤집은 후 다시 index를 0으로 설정하여 차례대로 출력
data_health.loc[::-1].reset_index(drop=True).head(10)
```

```python
#컬럼(열) 순서를 반대로 뒤집기
data_health.loc[:, ::-1].head()
```

```python
#결측 데이터 갯수 확인
data_health.isnull().sum()
```

```python
#중복 데이터 확인
data_health.duplicated().sum()
```

```python
#특정 행 삭제
data_health = data_health.drop([data_health.index[3], data_health.index[5]])
data_health
```

```python
#reset_insex를 이용해 인덱스 사이의 간격을 일정하게 유지하도록 갱신한다.
#drop=True로 기존의 인덱스는 없앤다. inplace=True로 적용한 것을 변수에 바로 적용한다.
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
#특정 컬럼의 값에 조건식을 설정하여 해당하는 행 모두 삭제
remove_index = data_health[data_health['WEIGHT'] == 60].index
data_health = data_health.drop(remove_index)
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
# 2개 이상의 조건문을 기술하여 and(&) 또는 or(|)로 적용
remove_index = data_health[(data_health['WEIGHT'] == 90) | (data_health['HEIGHT'] == 185)].index
data_health = data_health.drop(remove_index)
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
# append를 이용해 새로운 행 추가
new_list = [(170, 80, 70, 115, 14), (180, 80, 75, 113, 12)]
new_data = pd.DataFrame(new_list, columns=data_health.columns, index=[1200, 1221])
data_health = data_health.append(new_data)
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
# loc를 이용해 새로운 행 추가
data_health.loc[1433] = [175, 70, 77, 119, 12.6]
data_health.reset_index(drop=True, inplace=True)
data_health
```

```python
# 특정 컬럼(HEIGHT)을 기준으로 그룹화하고 각각 특정 컬럼(WEIGHT) 값의 평균 계산
data_health.groupby(['HEIGHT'])['WEIGHT'].mean()
```

```python
# 특정 컬럼(HEIGHT)을 기준으로 그룹화하고 각각 특정 컬럼(WEIGHT) 값의 수치 계산
data_health.groupby(['HEIGHT'])['WEIGHT'].agg(['mean', 'sum', 'count'])
```

```python
# 특정 컬럼(HEIGHT)을 기준으로 그룹화하고 각각 특정 컬럼(WEIGHT) 값으로 describe
data_health.groupby(['HEIGHT'])['WEIGHT'].describe()
```

```python
# index번호를 이용해 범위내의 행 삭제
data_health = pd.read_csv("health.csv", index_col=0)
data_health3 = data_health.drop(data_health.index[0:200])
data_health3
```

```python
data_health2 = data_health.drop(data_health.index[200:1500])
```

```python
# append를 이용해 2개의 데이터프레임을 세로방향으로 통합
data_health4 = data_health2.append(data_health3, ignore_index=True)
data_health4
```

```python
# concatdmf 이용해 2개의 데이터프레임을 세로방향으로 통합
data_health5 = pd.concat([data_health2, data_health3], axis=0)
data_health5
```
